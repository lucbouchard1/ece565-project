```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w = np.linspace(-5, 5, 100)

for n in [10, 20, 50]:
    plt.plot(w, (1 + np.exp(w))**2 / (n*np.exp(w)), label="Num Data Points = " + str(n))

plt.xlabel("w")
plt.ylabel("Variance in Estimate of w")
plt.legend()
plt.savefig('out.png')
```
